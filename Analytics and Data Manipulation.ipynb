````python
import seaborn as sns
````

````python
# Create a lmplot to color points by "class" and suppress the trend line
lm_plot = sns.lmplot(
    data=mpg,
    x="displ",
    y="hwy",
    ci=None,
    hue="class",  # Color points by the "class" variable
    fit_reg=False,  # Suppress the trend line
    scatter_kws={"s": 10},
    lowess = True,
    x_jitter=0.5,
    y_jitter=0.5
).set(
    title="Fuel consumption by engine displacement",
    xlabel="Engine displacement",
    ylabel="Fuel consumption"
).tight_layout()

# Create regplot
_ = sns.regplot(
    data=mpg,
    x="displ",
    y="hwy",
    ci=None,
    scatter=False,  # Suppress plotting the points
    line_kws = {"color": "black"}, # Color of the trend line 
    scatter_kws={"s": 10},
    lowess = True,
    x_jitter=0.5,
    y_jitter=0.5,
    ax=lm_plot.ax  # overlay the new plot to the previous one, rather than creating a new plot
)






````

```
ggplot(data = diamonds, mapping = aes(x = color, fill = clarity)) +
```

Hint: [this stackoverflow](https://stackoverflow.com/questions/49531655/issue-using-qualitative-brewer-palettes-in-plotnine) discussion might come in handy to set the palette colors. Also, note that the variable `clarity` is actually an ordered categorical variable (to see this, read the docs in R for the `diamonds` dataset with `? diamonds`). We can create a `pandas` equivalent of an R factor by using the `pandas.Categorical` class (see the docs [here](https://pandas.pydata.org/docs/reference/api/pandas.Categorical.html)). To get the same ordering that we got in R for the dodged bars, you will need to transform the `clarity` variable in the `diamonds` dataset that you read with `pandas` to a `pandas.Categorical` variable.

### Import Diamonds dataset

````python
diamonds = pd.read_csv("diamonds.csv")
````

````python
diamonds.describe()
````

````python
diamonds.head()
````

````python
#!pip install plotnine
````

````python
from plotnine import ggplot, aes, geom_bar, scale_fill_brewer,geom_jitter, geom_smooth, ggtitle, labs
````

````python
# Define the order of clarity levels to refer to it as categories
clarity_order = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]

# "clarity" variable into a pandas.Categorical
diamonds["clarity"] = pd.Categorical(diamonds["clarity"], categories=clarity_order, ordered=True)

# Plot using a different color palette
plot = (ggplot(diamonds, aes(x="color", fill="clarity"))
        + geom_bar(position="dodge")
        + scale_fill_brewer(type= "qual", palette="Set2")  
        + labs(x="Color", y="", fill="Diamond clarity", title="Clarity distribution by color"))
print(plot)


````

````python
dinners = pd.read_csv("dinners.csv")
````

````python
drinks = pd.read_csv("drinks.csv")
````

````python
food = pd.read_csv("food.csv")
````

````python
persons = pd.read_csv("persons.csv")
````

````python
# left joins
dinners_explicit = dinners.merge(drinks, left_on="drink_id", right_on="item_id", how="left") \
    .rename(columns={"price": "drink_price", "item_name": "drink"}) \
    .merge(food, left_on="food_id", right_on="food_id", how="left") \
    .rename(columns={"price": "food_price", "name": "food"}) \
    .merge(persons, left_on="person_id", right_on="id", how="left") \
    .drop(columns=["food_id", "person_id"])

dinners_explicit.head(15)


````

````python
# Group by 'drink'
top2_drinks = dinners_explicit.groupby('drink').size().reset_index(name='count')
top2_drinks = top2_drinks.sort_values(by='count', ascending=False)

# Remove rows with NaN values
top2_drinks = top2_drinks.dropna()

# Select the top 2 rows
top2_drinks = top2_drinks.head(2)
top2_drinks.head()
````

````python
# Group by 'food'
top2_food = dinners_explicit.groupby('food').size().reset_index(name='count')
top2_food = top2_food.sort_values(by='count', ascending=False)

# Remove rows with NaN values
top2_food = top2_food.dropna()

# Select the top 2 rows
top2_food = top2_food.head(2)

# Display the result
top2_food.head()
````
